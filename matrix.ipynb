```python
import numpy as np
```

Translation matrix

```python
def translation(x, y, z):

    t_matrix = np.identity(4)

    t_matrix[0,3] = x
    t_matrix[1,3] = y
    t_matrix[2,3] = z

    return t_matrix
```

Rotation Matrix

Rotation x - yaw

```python
def create_rotation_matrix_x(angle_deg):

    rad = np.radians(angle_deg)
    c = np.cos(rad)
    s = np.sin(rad)
    
    rx_matrix = np.identity(4)
    
    rx_matrix[1, 1] = c
    rx_matrix[1, 2] = -s
    rx_matrix[2, 1] = s
    rx_matrix[2, 2] = c
    
    return rx_matrix
```

Rotation y - pitch

```python
def create_rotation_matrix_y(angle_deg):

    rad = np.radians(angle_deg)
    c = np.cos(rad)
    s = np.sin(rad)
    
    ry_matrix = np.identity(4)

    ry_matrix[0, 0] = c
    ry_matrix[0, 2] = s
    ry_matrix[2, 0] = -s
    ry_matrix[2, 2] = c
    
    return ry_matrix
```

Rotation z - roll

```python
def create_rotation_matrix_z(angle_deg):

    rad = np.radians(angle_deg)
    c = np.cos(rad)
    s = np.sin(rad)

    rz_matrix = np.identity(4)

    rz_matrix[0, 0] = c
    rz_matrix[0, 1] = -s
    rz_matrix[1, 0] = s
    rz_matrix[1, 1] = c
    
    return rz_matrix
```

Concatenation of Rotation

```python
def calculate_concatennation_r(rx, ry, rz):
    
    cr_matrix = rz @ ry @ rx

    return cr_matrix
```

Transformation Matrix

```python
def transformation_matrix(crm, tm):

    tr_matrix = np.identity(4)

    tr_matrix[0:3, 0:3] = crm[0:3, 0:3]
    
    tr_matrix[0:3, 3] = tm[0:3, 3]
    
    return tr_matrix
```

Example

```python
tm = translation(2,3,5)

rx = create_rotation_matrix_x(34)
ry = create_rotation_matrix_y(54)
rz = create_rotation_matrix_z(123)

crm = calculate_concatennation_r(rx, ry, rz)

tr_m = transformation_matrix(crm, tm)

print(tr_m)
```

Inverse transformation

```python
print(np.linalg.inv(tr_m)) 
```
